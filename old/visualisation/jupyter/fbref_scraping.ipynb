```python
from matplotlib import pyplot as plt
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from io import StringIO
import os
import time
```

Get stats

```python
standings_url = "https://fbref.com/en/comps/9/Premier-League-Stats"
data = requests.get(standings_url)
soup = BeautifulSoup(data.text, features="lxml")
standings_table = soup.select("table.stats_table")[0]
```

```python
standings_table
```

```python

# links store href of each team
links = standings_table.find_all("a")
links = [l.get("href") for l in links]
links = [l for l in links if "/squads/" in l]

team_urls = [f"https://fbref.com{l}" for l in links]
```

```python
team_urls
```

```python
team_url = "https://fbref.com/en/squads/19538871/Manchester-United-Stats"

data = requests.get(team_url)
squad_df = pd.read_html(StringIO(data.text), match="Standard Stats")[0]
squad_df = squad_df.droplevel(level=0, axis=1)

# Get team name
team_name = team_url.split("/")[-1].replace("-Stats", "").replace("-", " ")
# Add a column team_name
squad_df["team"] = team_name

# Change index to lower case
squad_df.columns = [c.lower() for c in squad_df.columns]


time.sleep(1)

```
