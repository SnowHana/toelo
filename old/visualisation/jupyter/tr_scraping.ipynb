```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from bs4 import BeautifulSoup

```

```python

# Set up Selenium WebDriver
options = Options()
options.headless = True  # Run without opening a browser window
driver = webdriver.Firefox(options=options)

# Load the page
url = "https://www.transfermarkt.com/premier-league/transfers/wettbewerb/GB1/saison_id/2022"
driver.get(url)

# Wait for JavaScript to execute
# (You may need to adjust the waiting time based on the page load speed)
driver.implicitly_wait(5)  # Wait for 10 seconds

# Get the page source after JavaScript has executed
page_source = driver.page_source

# Close the WebDriver
driver.quit()

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(page_source, "html.parser")

# Now you can continue with parsing the content using BeautifulSoup
```
